```
# IMPORT STATEMENTS
# (None)

# FUNCTION DEFINITIONS
def verify_pin():
    while True:
        pin = input("Enter PIN: ")
        if pin == "1234":
            print("PIN Verified.")
            return True
        else:
            print("Incorrect PIN. Try again.\n")


def withdraw(balance, amount):
    if amount <= balance:
        balance -= amount
        print("Withdrawal successful.")
        return balance
    else:
        print("Insufficient funds.")
        return balance


# PROGRAM LOGIC
current_balance = 500

verify_pin()

amount = float(input("Enter withdrawal amount: "))

current_balance = withdraw(current_balance, amount)

print(f"Current Balance: ${current_balance}")
```
